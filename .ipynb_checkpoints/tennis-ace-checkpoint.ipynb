### <font color=DarkMagenta> Correlation Analysis

The correlation coefficient is a statistical measure of the strength of the relationship between the relative movements of two variables, in our example is the pairing of the DataFrame columns to measure the strength of the relationship between the two columns values. (See Coefficient of Correlation Matrix DataFrame).<br>
The calculated correlation values range between -1.0 and 1.0.<br>
A correlation of -1.0 shows a perfect negative correlation, while a correlation of 1.0 shows a perfect positive correlation.<br> 
A correlation of 0.0 shows no linear relationship between the movement of the two variables.<br>

The scatter plot is usefull to visually identify relationships between a first and a second entries of paired data.
- If it seems that the points follow a linear pattern well, a high linear correlation may exist between the paired date.  
- If it seems that the data do not follow a linear pattern, a no linear correlation may exist between the paired date.  
- If the data somewhat follow a linear path, a moderate linear correlation may exist between the paired date.

#### <font color=MediumBlue>Using Real Data to predict how the features affect winnings

In our single feature linear regressions, we will use the positive features ```BreakPointsOpportunities``` and ```Aces``` data, and the negative features ```BreakPointsFaced```, ```DoubleFaults``` data as a individual sets of variables and pair each of them with the ```Winnings``` outcome data.

````python
# Features name list
four_best_features = ['BreakPointsOpportunities','Aces' , 'BreakPointsFaced', 'DoubleFaults']
# Winnings list
winnings = man_tennis[['Winnings']]
````

````python
# Visuals relationships matrix scatter plots and winnings predictions (line of best fit)
plt.figure(figsize=(10, 12))
plt.subplots_adjust(wspace=0.55, hspace=0.5)
# Loops trought features
for i in range(len(four_best_features)):
    plt.subplot(2 , 2, 1+i) 
    plt.scatter(winnings, man_tennis[four_best_features[i]],  c='midnightblue', alpha=0.25)
    plt.xlabel(four_best_features[i], fontsize=13)
    plt.ylabel('Winnings', fontsize=17)
    plt.title(f'r = {round(man_tennis[four_best_features[i]].corr(man_tennis.Winnings), 2)}', fontsize=16)
    
    #------ Predicts winnings --------
    
    # Transforms the pd.series into a numpy array with 1 column 1 row to be use by .fit()
    feature = man_tennis[four_best_features[i]].to_numpy().reshape(-1, 1)
    
    # Linear Regression model
    reg = LinearRegression()
    # Line best fit
    # Finds the best coefficient 'm' and best y intercept relative to our feature and outcome 
    reg.fit(feature, winnings)
    # Predicts winnings relative to feature
    predict_winnings = reg.predict(feature)
    
    ax = plt.plot(predict_winnings, man_tennis[four_best_features[i]] , c='r')
    plt.legend(ax, ['Winnings Prediction'])

plt.savefig('graph/best_4_features_vs_winnings.png') 
plt.show()
````

Base on the real data, the predicted winnings amounts increase as the values of the features increase.

#### <font color=MediumBlue>Test Sets and Training Sets

For our project we want to test how well our four feature-winnings simple regression models score in predicting winnings.
-	We need to split our data set in training sets and test sets.
-	We want to train our models using the training sets.
-	And score the models, how well models do at predicting winnings using the test sets.
  
The model scores are the models coefficient of determination, <u style="font-size:11px">$R^2$</u>, for our example, is the coefficient of determinations of the winnings test set relative to each feature test set. 

"In statistics, the coefficient of determination, is the proportion of the variance in the dependent variable that is predictable from the independent variable(s).
It is a statistic used in the context of statistical models whose main purpose is either the prediction of future outcomes or the testing of hypotheses, on the basis of other related information. It provides a measure of how well observed outcomes are replicated by the model, based on the proportion of total variation of outcomes explained by the model."<a href="https://en.wikipedia.org/wiki/Coefficient_of_determination"> Coefficient of Determination<a>

````python
# Creates a DataFrame to store models coefficient of determination 
one_feature_scores = pd.DataFrame({'Feature':four_best_features, 'Score':float('NaN')})
# Creates an empty dictionary to store the features splits and models results
features_results = {}
````

````python
# Loops trough the four individual features and creates, trains and scores the models
for f in four_best_features:
    # Features values
    feature = man_tennis[[f]]
    
    # Splits the feature data set
    feature_train, feature_test, winnings_train, winnings_test = train_test_split(feature, winnings, train_size = 0.8)
    # Saves the splits results into the features results dictionary
    features_results[f'feature_train_{f}'] = feature_train
    features_results[f'feature_test_{f}'] = feature_test
    features_results[f'winnings_train_{f}'] = winnings_train
    features_results[f'winnings_test_{f}'] = winnings_test
                        
   
    # Creates a model base on linear Regression
    model = LinearRegression()
    # Saves model into the features results dictionary
    features_results[f'model_{f}'] = model
    
    # Find the best coefficient 'm' and best y intercept relative to our feature training sets and winnings training sets,
    # in other words, train the models using the training sets 
    model.fit(feature_train, winnings_train)
    
    # Scores the feature model using the feature test sets
    one_feature_scores.loc[one_feature_scores.Feature == f, 'Score'] = model.score(feature_test, winnings_test)
    
one_feature_scores.sort_values(by='Score', ascending=False).reset_index(drop=True).to_csv('data/scores_1f_winnings.csv')
one_feature_scores.sort_values(by='Score', ascending=False).reset_index(drop=True)
````

- The ```BreakPointsOpportunities``` feature, the number of times where the player could have won the service game of the opponent, is of the four features, our best option to predict ```Winnings```.
- The ```BreakPointsFaced``` feature, the number of times where the receiver could have won service game of the player, is of the four features, the second best option to predict ```Winnings```.
- The ```DoubleFaults``` feature, the number of times player missed both first and second serve attempts, is of the four features, the third best option to predict ```Winnings```.
- The feature ```Aces```, the number of serves by the player where the receiver does not touch the ball, is of the four features, the least best option to predict ```Winnings```.

#### <font color=MediumBlue>Models predictions vs actual data

Using our four regression models, we want to predict winnings relative to each feature, and compare it to the actual winnings by using a scatter plot graphs and calculating the coefficient of correlation between the predicted winnings and the actual winnings relative to each feature.   

````python
# Creates a new DataFrame to store the predicted winnings-actual winnings coefficient of correlation relative to each feature 
one_feature_winnings_vs_predictions_corr = pd.DataFrame({'Feature':four_best_features, 'coeff_r':float('NaN')})
````

````python
# Creates a figure
plt.figure(figsize=(19, 5))
plt.subplots_adjust(wspace=0.35)

# Loops trough the four individual features, creates scatters plot, calacutes and saves the coefficient of correlations
for i in range(len(four_best_features)):
    
    # Saves the model, feature test set and winnings test set into variables
    model = features_results[f'model_{four_best_features[i]}']
    feature_test = features_results[f'feature_test_{four_best_features[i]}']
    winnings_test = features_results[f'winnings_test_{four_best_features[i]}']
   
    # Makes predictions from the test data using the regression model
    predicted_winnings = model.predict(feature_test)
    
    #------ coefficient of correlation --------
    
    # Creates a DataFrame to store predicted winnings relative to the feature and actual winnings
    winnings_sets = pd.DataFrame()
    winnings_sets['predicted_winnings'] = predicted_winnings.reshape(len(predicted_winnings),)  
    winnings_sets['actual_winnings'] = winnings_test.Winnings.values
    
    # Creates a coefficient of correlation matrix DataFrame 
    winnings_sets_corr = winnings_sets.corr().reset_index()
    
    # Saves the predicted winnings-actual winnings coefficient of correlation realative the feature 
    one_feature_winnings_vs_predictions_corr.loc[one_feature_winnings_vs_predictions_corr.Feature == \
                                                 f'{four_best_features[i]}', 'coeff_r'] = \
                                                            winnings_sets_corr.actual_winnings[0]  
    
    #------ Scatter plot --------
    
    # Visuals model predicted winnings vs actual winnings 
    plt.subplot(1 , len(four_best_features), 1+i) 
    plt.scatter(winnings_test, predicted_winnings,  c='midnightblue', alpha=0.25)
    plt.xlabel('Actual Winnings', fontsize=16)
    if four_best_features[i] == 'BreakPointsOpportunities': 
        plt.ylabel('Winnings', fontsize=17)
    plt.title(f'{four_best_features[i]}\n$R^2$={round(one_feature_scores.Score[i], 2)}', fontsize=15)
    
    # Text-box diplaying the coefficient of determination, R^2
    plt.text(0, np.max(predicted_winnings), 
             f'r={round(winnings_sets_corr.actual_winnings[0], 2)}', 
             fontsize=14, 
             horizontalalignment='left',
             verticalalignment='center')
    
plt.savefig('graph/one_f_predicted_vs_winnings.png') 
plt.show()
````

### <font color=DarkMagenta>Multiple Linear Regression

Multiple Linear Regression uses two or more independent variables to predict the values of the dependent variable.

In the process of refactoring  the code, I created the ```reg()``` function.

The ```reg()``` funtion function:
- Takes the arguments:
    - ```features_name```, string type, the name of the single or multiple features.
    - ```features_values```, DataFrame type, independent variables, features values combinaisons.
    - ```winnings```, DataFrame type, dependent variable, ```Winnings``` values combinaisons.
    - ```df_coeff```, DataFrame type, storing the the coefficient of determination and coefficient of correlation results.
    - ```relevent_score```, float type, set the coefficient of determination value of relevence.
<br><br>   
- splits, fits and predicts winnings.<br> 
- computes the features vs winnings coefficient of determination and the predicted winnings vs winnings coefficient of correlation, and save coefficient in DataFrame
<br><br>
- returns the number of the scores equal to or over the ```relevent_score``` value.

````python
# Creates an empty dictionary to store the test and predicted winnings 
features_results={}

def reg(features_name, features_values, winnings, df_coeff, relevent_score):
    
    # Splits the features data set
    features_train, features_test, winnings_train, winnings_test = train_test_split(features_values, 
                                                                                    winnings, 
                                                                                    train_size = 0.80)

    features_results[f'{features_name}_winnings_test'] = winnings_test
    
  
    # Creates a model base on linear Regression
    model = LinearRegression()
    # Trains the models using the training sets 
    model.fit(features_train, winnings_train)
    # Makes predictions from the test data using the regression model
    predicted_winnings = model.predict(features_test)
    
    # Saves predicted_winnings
    features_results[f'{features_name}_predicted_winnings'] = predicted_winnings
    

    #------ the coefficient of determination --------

    score = model.score(features_test, winnings_test)
    # Saves score
    df_coeff.loc[df_coeff.Features == features_name, 'Score'] = score
    
    #------ coefficient of correlation --------

    # Creates a DataFrame to store predicted winnings relative to the two features and actual winnings
    winnings_sets = pd.DataFrame()

    winnings_sets['predicted_winnings'] = predicted_winnings.reshape(len(predicted_winnings),) 
    winnings_sets['actual_winnings'] = winnings_test.Winnings.values

    # Creates a coefficient of correlation matrix DataFrame 
    winnings_sets_corr = winnings_sets.corr().reset_index()
    
    # Saves the predicted winnings-actual winnings coefficient of correlation realative to the two features combination
    df_coeff.loc[df_coeff.Features == features_name, 'coeff_r'] = winnings_sets_corr.actual_winnings[0]
    
    # Relevent coefficient of determination
    if score >= relevent_score:
        return 1
    else:
        return 0
````

#### <font color=MediumBlue>Two Features Linear regression

We want to predict winnings using two features combination.

For our dependent variables we will use: 
- The ```Winnings``` outcome: number of matches won in a year.

For our independent variables we will use:<br>

- Positive two features combinations using:
    - ```Aces```: number of serves by the player where the receiver does not touch the ball
    - ```BreakPointsOpportunities```: number of times where the player could have won the service game of the opponent
    - ```FirstServePointsWon```: % of first-serve attempt points won by the player
    - ```SecondServePointsWon```: % of second-serve attempt points won by the player
    - ```BreakPointsSaved```: % of the time the player was able to stop the receiver from winning service game when they had the chance
    - ```ServiceGamesWon```: total number of games where the player served and won
    - ```TotalServicePointsWon```: % of points in games where the player served that they won
    - ```FirstServeReturnPointsWon```: % of opponents first-serve points the player was able to win
    - ```SecondServeReturnPointsWon```: % of opponents second-serve points the player was able to win
    - ```BreakPointsConverted```: % of the time the player was able to win their opponent’s service game when they had the chance
    - ```ReturnGamesWon```: total number of games where the player’s opponent served and the player won
    - ```ReturnPointsWon```: total number of points where the player’s opponent served and the player won
    - ```TotalPointsWon```: % of points won by the player
<br><br>
- Negative two features combinations using:
    - ```BreakPointsFaced```: number of times where the receiver could have won service game of the player
    - ```DoubleFaults```: number of times player missed both first and second serve attempts

In the process of refactoring  the code, I created the ```two_features_reg()``` function.

The ```two_features_reg()``` function:
- Takes the arguments:
    - ```two_features_name```, list type, two features combination name list.
    - ```df_coeff```, DataFrame type, storing the the coefficient of determination and coefficient of correlation results.
    - ```relevent_score```, float type, set the coefficient of determination value of relevence.
<br><br> 
- calls the function reg():
    - splits, fits and predicts winnings.<br> 
    - computes the features vs winnings coefficient of determination and the predicted winnings vs winnings coefficient of correlation, and save coefficient in DataFrame.
    - returns the number of the scores equal to or over the ```relevent_score``` value.
<br><br>
- returns:
    - the number of rows needed for a 3 columns ploting grid.
    - returns the number of the scores equal to or over the ```relevent_score``` value.

````python
def two_features_reg(two_features_names, df_coeff, relevent_score):
    
    # must relevent coefficients of determination counter 
    counter=0
    
    # Loops trough features, calacutes and saves the coefficients of determination and correlation
    # Sums must relevent coefficients of determination
    for i in range(len(two_features_names)-1):

        for j in range(i+1, len(two_features_names)):

            features_name = f'{two_features_names[i]}_{two_features_names[j]}'

            # Features values
            two_features_values = man_tennis[[f'{two_features_names[i]}', f'{two_features_names[j]}']]

            # Linear regression model computing and results saving function
            # sums the must relevent coefficients of determination
            counter += reg(features_name, two_features_values, winnings, df_coeff, relevent_score)

    # Rows counter, number of rows needed for a 3 columns ploting grid
    if counter%3 == 0:
        rows = counter/3
    else:
        rows = (counter+3-(counter%3))/3
        
    return int(rows), counter 
````

In the process of refactoring  the code, I created the ```two_features_reg_plot()``` function.

The ```two_features_reg_plot()``` function:
- Takes the arguments:
    - ```two_features_name```, list type, two features combination name list.
    - ```df_coeff```, DataFrame type, storing the the coefficient of determination and coefficient of correlation results.
    - ```rows```, integer type, the number of rows.
    - ```relevent_score```, float type, set the coefficient of determination value of relevence.
<br><br> 
- outputs a predicted  winnings vs actual winnings 3 columns scatter plots grid from the most relevent two features combinations.

````python
def two_features_reg_plot(features_names, df_coeff, rows, relevent_score):  
    # Plot index counter
    k=1 
    # Loops trough features, creates  scatter plots for predicted winnings vs test winnings
    # with the must relevent coefficients of determination   
    for i in range(len(features_names)-1):

        for j in range(i+1, len(features_names)):

            two_features_name = f'{features_names[i]}_{features_names[j]}'
            
            #------ the coefficient of determination ------
            score = df_coeff.Score[df_coeff.Features == two_features_name].values

            # Plot if score is relevent
            if score >= relevent_score:
                
                #------ coefficient of correlation --------
                coeff_r = df_coeff.coeff_r[df_coeff.Features == two_features_name].values

                # actual and predicted winnings
                winnings_test = features_results[f'{two_features_name}_winnings_test']
                predicted_winnings = features_results[f'{two_features_name}_predicted_winnings']

                #------ Scatter plot --------
                # Visuals model predicted winnings vs actual winnings 
                plt.subplot(rows, 3, k) # Beautiful
                plt.scatter(winnings_test, predicted_winnings,  c='midnightblue', alpha=0.25)
                plt.xlabel('Actual Winnings', fontsize=12) 
                plt.ylabel('Winnings', fontsize=13)
                plt.title(f'{features_names[i]}\n{features_names[j]}\n$R^2$={round(float(score), 2)}', fontsize=11)
                #Text-box diplaying the coefficient of correlation, r
                plt.text(0, np.max(predicted_winnings), 
                                   f'r={round(float(coeff_r), 2)}', 
                                   fontsize=12, 
                                   horizontalalignment='left',
                                   verticalalignment='top')
                k+=1
````

````python
# sets relevent coefficient of determination value
relevent_score = 0.86
# Plot index counter
k=1
# must relevent coefficients of determination counter 
counter=0


for i in range(len(positive_features_names)-1):

    for j in range(i+1, len(positive_features_names)):
        
        for g in range(j+1, len(positive_features_names)):
            
             for l in range(g+1, len(positive_features_names)):
            
            
                features_name = f'{positive_features_names[i]}_{positive_features_names[j]}_{positive_features_names[g]}_{positive_features_names[l]}'

                # Features values
                four_features_values = man_tennis[[f'{positive_features_names[i]}', 
                                                    f'{positive_features_names[j]}', 
                                                    f'{positive_features_names[g]}',
                                                    f'{positive_features_names[l]}']]


                # Linear regression model computing and results saving function
                # sums the must relevent coefficients of determination
                counter += reg(features_name, four_features_values, winnings, positive_four_features_coeff, relevent_score)

if counter == 0:
    print('All the positive four features linear regression models have score results under the revelent score of ' + str(relevent_score))
else:

    # Rows counter, number of rows needed for a 3 columns ploting grid
    if counter%3 == 0:
        rows = counter/3
    else:
        rows = (counter+3-(counter%3))/3


    # Creates a figure
    plt.figure(figsize=(15, 27))
    plt.subplots_adjust(wspace=0.4, hspace=0.6)  

    for i in range(len(positive_features_names)-1):

        for j in range(i+1, len(positive_features_names)):

            for g in range(j+1, len(positive_features_names)):

                 for l in range(g+1, len(positive_features_names)):

                    features_name = f'{positive_features_names[i]}_{positive_features_names[j]}_{positive_features_names[g]}_{positive_features_names[l]}'

                    #------ the coefficient of determination ------
                    score = positive_four_features_coeff.Score[positive_four_features_coeff.Features == features_name].values


                    # Plot if score is relevent
                    if score >= relevent_score:

                        #------ coefficient of correlation --------
                        coeff_r = positive_four_features_coeff.coeff_r[positive_four_features_coeff.Features == features_name].values

                        # actual and predicted winnings
                        winnings_test = features_results[f'{features_name}_winnings_test']
                        predicted_winnings = features_results[f'{features_name}_predicted_winnings']

                        #------ Scatter plot --------
                        # Visuals model predicted winnings vs actual winnings 
                        plt.subplot(int(rows), 3, k) 
                        plt.scatter(winnings_test, predicted_winnings,  c='midnightblue', alpha=0.25)
                        plt.xlabel('Actual Winnings', fontsize=12) 
                        plt.ylabel('Winnings', fontsize=13)
                        plt.title(f'{positive_features_names[i]}\n{positive_features_names[j]}\n{positive_features_names[g]}\n{positive_features_names[l]}\n$R^2$={round(float(score), 2)}', fontsize=11)
                        #Text-box diplaying the coefficient of correlation, r
                        plt.text(0, np.max(predicted_winnings), 
                                           f'r={round(float(coeff_r), 2)}', 
                                           fontsize=12, 
                                           horizontalalignment='left',
                                           verticalalignment='top')

                        k+=1

    plt.savefig('graph/positive_four_f_predicted_vs_winnings.png')
    plt.show()
````

For our independent variables we will use:<br>

    - ```Aces```: number of serves by the player where the receiver does not touch the ball
    - ```BreakPointsOpportunities```: number of times where the player could have won the service game of the opponent
    - ```FirstServePointsWon```: % of first-serve attempt points won by the player
    - ```SecondServePointsWon```: % of second-serve attempt points won by the player
    - ```BreakPointsSaved```: % of the time the player was able to stop the receiver from winning service game when they had the chance
    - ```ServiceGamesWon```: total number of games where the player served and won
    - ```TotalServicePointsWon```: % of points in games where the player served that they won
    - ```FirstServeReturnPointsWon```: % of opponents first-serve points the player was able to win
    - ```SecondServeReturnPointsWon```: % of opponents second-serve points the player was able to win
    - ```BreakPointsConverted```: % of the time the player was able to win their opponent’s service game when they had the chance
    - ```ReturnGamesWon```: total number of games where the player’s opponent served and the player won
    - ```ReturnPointsWon```: total number of points where the player’s opponent served and the player won
    - ```TotalPointsWon```: % of points won by the player
